```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('../data/clean_data.csv')
print(f'clean data loaded Shape:{df.shape}')
```

```python
#checking current columns

numeric_cols=df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols=df.select_dtypes(include=['object']).columns.tolist()
print(f'Numeric columns :{len(numeric_cols)}')
print(f'Categorical columns :{len(categorical_cols)}')


                                  
```

```python
# Create New Features:

#total square Footage
df['TotalSF']=df['Total Bsmt SF']+df['1st Flr SF'] + df['2nd Flr SF']

#House Age
df['HouseAge']=df['Yr Sold'] - df['Year Built']
#Was it remodeled?
df['IsRemodeled']=(df['Year Remod/Add']!=df['Year Built']).astype(int)

#total bathrooms
df['TotalBath']=(df['Full Bath'] +
                 df['Half Bath'] *0.5+
                 df['Bsmt Full Bath']+
                 df['Bsmt Half Bath'] *0.5)
print('New features created')
print('fNew shape:{df.shape}')
print('\nNew columns added:')
print('-TotalSf: TOtal square footage')
print('-Houseing Age: Age of house when sold')
print('-IsRemoved: 1 if remodeled, 0 if not')
print('-TotalBath : Total bathrooms combined')

                 
```

```python
#Log Transform SalePrice
#salePrice is skewed : Log transform makes if  normal
df['SalePrice_log']=np.log(df['SalePrice'])

#compare before and after
fig,axes = plt.subplots(1,2,figsize=(12,4))

axes[0].hist(df['SalePrice'],bins=50,color='steelblue',edgecolor='white')
axes[0].set_title('SalePrice-Original(Skewed)')
axes[0].set_xlabel('Sale Price($)')

axes[1].hist(df['SalePrice_log'],bins=50,color='coral',edgecolor='White')
axes[1].set_title('SalePrice - After Log(Normal)')
axes[1].set_xlabel('Log Sale Price')

plt.tight_layout()
plt.show()

print('Log transform applied to SalePrice')
```

```python
# One-hot encoding converts text categories to numbers


df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print(f'Shape before encoding: {df.shape}')
print(f'Shape after encoding : {df_encoded.shape}')
print('Categorical columns encoded!')
```

```python
#check new features vs price
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('New Features vs SalePrice', fontsize=16, fontweight='bold')

# TotalSF vs Price
axes[0, 0].scatter(df['TotalSF'], df['SalePrice'], alpha=0.4, color='steelblue')
axes[0, 0].set_title('Total Square Footage vs Price')
axes[0, 0].set_xlabel('Total SF')
axes[0, 0].set_ylabel('Sale Price ($)')

# HouseAge vs Price
axes[0, 1].scatter(df['HouseAge'], df['SalePrice'], alpha=0.4, color='coral')
axes[0, 1].set_title('House Age vs Price')
axes[0, 1].set_xlabel('House Age (years)')
axes[0, 1].set_ylabel('Sale Price ($)')

# TotalBath vs Price
sns.boxplot(x=df['TotalBath'].round(), y=df['SalePrice'], ax=axes[1, 0], palette='Blues')
axes[1, 0].set_title('Total Bathrooms vs Price')
axes[1, 0].set_xlabel('Total Bathrooms')
axes[1, 0].set_ylabel('Sale Price ($)')

# IsRemodeled vs Price
sns.boxplot(x=df['IsRemodeled'], y=df['SalePrice'], ax=axes[1, 1], palette='Greens')
axes[1, 1].set_title('Remodeled vs Not Remodeled')
axes[1, 1].set_xlabel('Is Remodeled (1=Yes, 0=No)')
axes[1, 1].set_ylabel('Sale Price ($)')

plt.tight_layout()
plt.show()
```

```python
# prepare final dataset
# Use log price as target
X = df_encoded.drop(['SalePrice', 'SalePrice_log'], axis=1)
y = df_encoded['SalePrice_log']

print(f'Features (X) shape : {X.shape}')
print(f'Target (y) shape   : {y.shape}')
print(f'\nData ready for modeling')
print(f'Total features: {X.shape[1]}')
```

```python
#saved prepare dara
df_encoded.to_csv('../data/prepared_data.csv',index=False)
print('Prepared Data Saved')
```
