```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 100)
plt.style.use('seaborn-v0_8')

print('libraries loaded')
```

```python

# load dataset
df=pd.read_csv('../data/AmesHousing.csv')

print(f'dataset loaded')
print(f' Rows :{df.shape[0]}')
print(f' Columns :{df.shape[1]}')
```

```python
df.head()
```

```python
#column names
print(df.columns.tolist())
```

```python
#datatypes
numeric_cols=df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols=df.select_dtypes(include=['object']).columns.tolist()

print(f'Numeric columns :{len(numeric_cols)}')
print(f'Categorical columns:{len(categorical_cols)}')
```

```python
# missing values
missing=df.isnull().sum()
missing_pct=(missing / len (df)) * 100

missing_df=pd.DataFrame({
    'Missing Count':missing,
    'Missing %': missing_pct.round(2)
})

missing_df=missing_df[missing_df['Missing Count']>0].sort_values('Missing %',ascending=False)
print(f'columns with missing values: {len(missing_df)}')
print()
print(missing_df)
    
```

```python
#salePrice summary
print(f'Mean: ${df["SalePrice"].mean():,.0f}')
print(f'Median: ${df["SalePrice"].median():,.0f}')
print(f'Min: ${df["SalePrice"].min():,.0f}')
print(f'Max: ${df["SalePrice"].max():,.0f}')
```

```python
# Visualizations:
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('House price EDA', fontsize=16, fontweight='bold')

# Plot 1: SalePrice Distribution
axes[0, 0].hist(df['SalePrice'], bins=50, color='steelblue', edgecolor='white')
axes[0, 0].set_title('SalePrice Distribution')
axes[0, 0].set_xlabel('Sale Price ($)')        
axes[0, 0].set_ylabel('Count')

# Plot 2: Log SalePrice
axes[0, 1].hist(np.log(df['SalePrice']), bins=50, color='coral', edgecolor='white')
axes[0, 1].set_title('Log(SalePrice) - Less Skewed')
axes[0, 1].set_xlabel('Log Sale Price')
axes[0, 1].set_ylabel('Count')

# Plot 3: Missing values bar chart
top_missing = missing_df.head(20)
axes[1, 0].barh(top_missing.index, top_missing['Missing %'], color='tomato')
axes[1, 0].set_title('Top 20 Columns with Missing Values')
axes[1, 0].set_xlabel('Missing %')

# Plot 4: Quality vs Price
qual_col = 'Overall Qual' if 'Overall Qual' in df.columns else 'OverallQual'
df.groupby(qual_col)['SalePrice'].median().plot(
    kind='bar', ax=axes[1, 1], color='mediumseagreen')   
axes[1, 1].set_title('Median SalePrice by Overall Quality')
axes[1, 1].set_xlabel('Overall Quality (1-10)')
axes[1, 1].set_ylabel('Median Sale Price ($)')

plt.tight_layout()
plt.show()

```

```python
#Top correlations
corr = df[numeric_cols].corr()['SalePrice'].drop('SalePrice')
top_corr = corr.abs().sort_values(ascending=False).head(15)

plt.figure(figsize=(10, 6))
top_corr.plot(kind='bar', color='steelblue')
plt.title('Top 15 Features Correlated with SalePrice')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(top_corr)
```
