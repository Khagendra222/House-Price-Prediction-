```python
#import libraries & Load Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('../data/AmesHousing.csv')
numerice_cols=df.select_dtypes(include=[np.number]).columns.tolist()
print(f' data loaded {df.shape[0]}rows, {df.shape[1]}columns')
```

```python
# scatter plot : sixe vs Price

plt.figure(figsize=(10,6))
plt.scatter(df['Gr Liv Area'],df['SalePrice'],alpha=0.4,color='steelblue')
plt.title('Living Area vs Sale Price')
plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Peice($)')
plt.tight_layout()
plt.show()

print('Bigger houses=higher price(mostly)')
```

```python
# Scatter Plot : Year Built vs Price

plt.figure(figsize=(10,6))
plt.scatter(df['Year Built'],df['SalePrice'],alpha=0.4,color='coral')
plt.title('Year Built vs Sale Price')
plt.xlabel('Year Built')
plt.ylabel('Sale Price($)')
plt.tight_layout()
plt.show()
print(f'Insight: Newer houses tends to sell for more')
```

```python
#Box plot: Overall Quality vs Price

plt.figure(figsize=(12,6))
qual_col='Overall Qual' if 'Overall Qual' in df.columns else 'OverallQual'
sns.boxplot(x=qual_col,y='SalePrice',data=df,palette='Blues')
plt.title('Overall Quality vs Sale Price')
plt.xlabel('Overall Quality(1-10)')
plt.ylabel('Sale Price($)')
plt.tight_layout()
plt.show()
print('Insight: Quality rating is strongest predictor of price')
```

```python
#Boxplot:Neighborhood vs Price

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
plt.figure(figsize=(16,6))
neighborhood_order=df.groupby('Neighborhood')['SalePrice'].median().sort_values(ascending=False).index
sns.boxplot(x='Neighborhood',y='SalePrice',data=df,order=neighborhood_order,palette='coolwarm')
plt.title('Neighborhood vs Sale Price')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price($)')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
print('Insight: Location matters a lot- some neighborhoods are way more expensive')
           
```

```python
# Correlation Heatmap (Top 15 features)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

corr = df[numeric_cols].corr()
top_features = corr['SalePrice'].abs().sort_values(ascending=False).head(15).index

plt.figure(figsize=(14, 10))
sns.heatmap(
    df[top_features].corr(),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5
)
plt.title('Correlation Heatmap - Top 15 Features')
plt.tight_layout()
plt.show()

print(' Insight: Red = strong positive, Blue = strong negative correlation')
```

```python
# Distribution of Top 6 Features

top_6=corr['SalePrice'].abs().sort_values(ascending=False)[1:7].index
fig,axes=plt.subplots(2,3,figsize=(16,10))

for i,col in enumerate(top_6):
    row=i//3
    col_idx=i%3
    axes[row,col_idx].hist(df[col].dropna(),bins=30,color='steelblue',edgecolor='white')
    axes[row,col_idx].set_title(col)
    axes[row,col_idx].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot :Garage vs Price
garage_col='Garage Cars' if 'Garage Cars' in df.columns else 'GarageCars'

plt.figure(figsize=(10,6))
sns.boxplot(x=garage_col,y='SalePrice',data=df,palette='Greens')
plt.title('Garage Size vs Sale Price')
plt.xlabel('Garage Capacity(cars)')
plt.ylabel('Sale Price($)')
plt.tight_layout()
plt.show()

print('Insight: More garage space = higher price')
```
